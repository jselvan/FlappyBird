```python
from PIL import Image
import os

output_dir = "static/assets/icons"
os.makedirs(output_dir, exist_ok=True)

for skin_key, skin in ALL_SKINS.items():
    # Open images in order: backWing, body, frontWing
    back_wing = Image.open(skin["backWing"]).convert("RGBA")
    body = Image.open(skin["body"]).convert("RGBA")
    front_wing = Image.open(skin["frontWing"]).convert("RGBA")

    # Assume all images are the same size as body
    canvas = Image.new("RGBA", body.size)
    canvas.paste(back_wing, (0, 0), back_wing)
    canvas.paste(body, (0, 0), body)
    canvas.paste(front_wing, (0, 0), front_wing)

    # Save as webp using the skin_key as filename
    name = skin_key + ".webp"
    save_path = os.path.join(output_dir, name)
    canvas.save(save_path, "WEBP")
```
